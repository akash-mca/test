```python
from nltk.tokenize import TreebankWordTokenizer, sent_tokenize, WordPunctTokenizer, RegexpTokenizer
file_content = open('dataset.txt').read()
```

```python
tokenizer_wrd = TreebankWordTokenizer()
tokenizer_wrd.tokenize(file_content)
```

```python
tokenizer = WordPunctTokenizer()
tokenizer.tokenize(file_content)
```

```python
sent_tokenize(file_content)
```

```python
tokenizer = RegexpTokenizer("[\w']+")
tokenizer.tokenize(file_content)
```

```python
tokenizer = RegexpTokenizer('/s+' , gaps ="TRUE")
tokenizer.tokenize(file_content)
```
